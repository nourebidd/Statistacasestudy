```python
rev = df_last3.pivot_table(
    index=["Company", "Category"],
    columns="Year",
    values="Revenue",
    aggfunc="first"
).reset_index()

# Clean column names
rev.columns = ["company", "industry", "rev_2021", "rev_2022", "rev_2023"]

rev.head()

```

```python
optional_kpis = {
    "Net Income": "net_income",
    "Number of Employees": "employees",
    "Market Cap(in B USD)": "market_cap_b_usd"
}

for original, newname in optional_kpis.items():
    temp = df_last3.pivot_table(
        index="Company",
        columns="Year",
        values=original,
        aggfunc="first"
    )
    temp.columns = [f"{newname}_{y}" for y in temp.columns]
    rev = rev.merge(temp, left_on="company", right_index=True, how="left")

rev.head()
```

```python
import pandas as pd

# Load all 3 Forbes files
df20 = pd.read_excel("Forbes Global 2000 - 2020 — unified industry.xlsx")
df21 = pd.read_excel("Forbes Global 2000 - 2021 — unified industry.xlsx")
df22 = pd.read_excel("Forbes Global 2000 - 2022 — unified industry.xlsx")

# Select only necessary columns and rename for merging
def prep(df, year):
    temp = df.rename(columns={
        "Company": "company",
        "Country": "country",
        "Industry": "industry",
        "Sales": f"rev_{year}",
        "Profits": f"net_income_{year}",
        "Assets": f"assets_{year}",
        "Market_Value": f"market_cap_{year}"
    })
    return temp[["company","country","industry",
                 f"rev_{year}", f"net_income_{year}",
                 f"assets_{year}", f"market_cap_{year}"]]

df20c = prep(df20, 2020)
df21c = prep(df21, 2021)
df22c = prep(df22, 2022)

# MERGE on company name
merged = df20c.merge(df21c, on=["company","country","industry"], how="outer")
merged = merged.merge(df22c, on=["company","country","industry"], how="outer")

# Sort & take top 300 if needed
merged = merged.sort_values("company").reset_index(drop=True)
merged300 = merged.head(300)

merged300.head()
```
